```python
def find_elements_by_css_selector(selector) :
    try:
        a1 = driver.find_elements(by=By.CSS_SELECTOR, value=selector)
        return a1
    except:
        return None
def find_element_by_css_selector(selector) :
    try:
        a1 = driver.find_element(by=By.CSS_SELECTOR, value=selector)
        return a1
    except:
        return None
```

```python
# 전달된 태그로 부터 문자열 데이터를 추출해 반환하는 함수
def getText(tag):
    # 태그가 없다면
    if tag == None :
        return np.nan
    # 문자열을 가져온다.
    data = tag.text.strip()
    # 길이가 0 이라면...
    if len(data) == 0 :
        return np.nan
    return data
```
