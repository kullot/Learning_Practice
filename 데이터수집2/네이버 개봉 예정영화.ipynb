```python
import bs4
import requests
import pandas as pd
import numpy as np
import time
import os
from IPython.display import clear_output
import datetime
```

```python
# 요청 함수
def getSource(site) :
    # 해더 정보 셋팅
    header_info = {
        'User-Agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/104.0.0.0 Safari/537.36'
    }
    # 요청한다.
    response = requests.get(site, headers=header_info)
    return response.text
```

```python
# 한 페이지의 데이터를 가져오는 함수
def getData(source):
    # bs4 객체를 생성한다.
    soup = bs4.BeautifulSoup(source)
    
    # 데이터를 담을 딕셔너리
    data_dict = {
        '개봉예정일' : [],
        '제목' : [],
        '기대해요' : [],
        '기대안해요': [],
        '개요' : [],
        '감독' : [],
        '출연' : [],
        '상영시간' : [],
        '개봉일' : [],
        '영화코드' : []
    }
    # li_list1 = soup.select('div.day_t1')
    # for id_num1, li_tag1 in enumerate(li_list1):
    #     a = li_tag1.select_one('strong')
    #     data = getText(a).split('\n')[-1]
    #     print(data)
        
    # li 태그들을 가져온다.
    # li_list = soup.select('div.lst_wrap')
    li_list = soup.select('dl.lst_dsc')
    for id_num, li_tag in enumerate(li_list):
        # print(li_tag)
        # print('-------------------------')
        
        # 제목
        a1 = li_tag.select_one('dl > dt > a')
        data1 = getText(a1)
        
        # 기대해요
        a2 = li_tag.select_one('dd.star > dl > dd > div > em:nth-child(2)')
        data2 = getText(a2)
        
        # 기대인해요
        a3 = li_tag.select_one('dd.star > dl > dd > div > em:nth-child(4)')
        data3 = getText(a3)
        
        # 개요
        a4 = li_tag.select_one('dd:nth-child(2) > span.link_txt')
        data4 = getText(a4)
        if pd.notna(data4) :
            data4 = data4.replace('\r', '')
            data4 = data4.replace('\n', '')
            data4 = data4.replace('\t', '')
            data4 = data4.replace(' ', '')
        
        # 감독
        a5 = li_tag.select_one('dd:nth-child(3) > dl > dd:nth-child(4) > span')
        data5 = getText(a5)
        if pd.notna(data5) :
            data5 = data5.replace('\r', '')
            data5 = data5.replace('\n', '')
            data5 = data5.replace('\t', '')
            data5 = data5.replace(' ', '')
        
        # 출연
        a6 = li_tag.select_one('dd:nth-child(3) > dl > dd:nth-child(6) > span')
        data6 = getText(a6)
        if pd.notna(data6) :
            data6 = data6.replace('\r', '')
            data6 = data6.replace('\n', '')
            data6 = data6.replace('\t', '')
            data6 = data6.replace(' ', '')
        
        # 상영시간, 개봉일
        a7 = li_tag.select_one('dd:nth-child(3) > dl > dd:nth-child(2)')
        a8 = getText(a7).strip()
        if pd.notna(a8) :
            a8 = a8.replace('\r', '')
            a8 = a8.replace('\n', '')
            a8 = a8.replace('\t', '')
            a8 = a8.replace(' ', '').split('|')
            data7 = a8[-2][:-1]
            data8 = a8[-1][:-2]

        # 영화코드
        a9 = li_tag.select_one('dt > a')
        data9 = a9.attrs['href'].split('=')[-1]
        
        # 담는다.
        
        data_dict['제목'].append(data1)
        data_dict['기대해요'].append(data2)
        data_dict['기대안해요'].append(data3)
        data_dict['개요'].append(data4)
        data_dict['감독'].append(data5)
        data_dict['출연'].append(data6)
        data_dict['상영시간'].append(data7)
        data_dict['개봉일'].append(data8)
        data_dict['영화코드'].append(data9)
    # dt_list = pd.to_datetime(data_dict['개봉일'])
    # dt_list = dt_list.dt.day_name()
    # print(dt_list)
    # data_dict['개봉예정일'] = 
            
        print(data1)
        print(data2)
        print(data3)
        print(data4)
        print(data5)
        print(data6)
        print(data7)
        print(data8)
        print(data9)
        print('-------------------------')
```

```python
# 전달된 태그로 부터 문자열 데이터를 추출해 반환하는 함수
def getText(tag):
    # 태그가 없다면
    if tag == None :
        return np.nan
    # 문자열을 가져온다.
    data = tag.text.strip()
    # 길이가 0 이라면...
    if len(data) == 0 :
        return np.nan
    return data
```
