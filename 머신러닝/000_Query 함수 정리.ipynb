```python
import pandas as pd
```

```python
df1 = pd.read_csv('./data/train.csv')
df1
```

```python
# 기본적으로 파이썬의 비교 연산자 사용이 가능하다.
a1 = df1.query('Age > 30')
a2 = df1.query('Age <= 30')
a3 = df1.query('CryoSleep == True')
display(a1)
display(a2)
display(a3)
```

```python
# 문자열 값과 비교하여 가져온다.
# 조건문자열에 문자열 값을 따옴표로 묶어줘야 한다.
df1.query('HomePlanet == "Earth"')
```

```python
# 따옴표로 묶어주지 않으면 컬럼 이름으로 인식한다.
# CryoSleep컬럼과 Transported컬럼의 값이 같은 것을 가져온다.
df1.query('CryoSleep == Transported')
```

```python
# and 조건
# HomePlanet이 Mars 이고 Age가 30보다 큰 데이터..
df1.query('HomePlanet == "Mars" and Age > 30')
```

```python
# or 조건
# HomePlanet이 Earth이거나 VIP가 True인 데이터
df1.query('HomePlanet == "Earth" or VIP == True')
```

```python
# not
# 주어진 조건 문자열의 부정
# HomePlanet이 Earth가 아닌 사람들
df1.query('not HomePlanet == "Earth"')
```

```python
# 변수에 잇는 겂을 사용한다.
a1 = 30

# 변수에 해당하는 부분은 @를 붙혀준다.
df1.query('Age > @a1')
```

```python
# in : 포함되어 있는가...
a1= ['Earth', 'Mars']
# HomePlanet의 값이 a1 리스트에 있는 것과 같은 것들만 가져온다.
a2 = df1.query('HomePlanet in @a1')
a2['HomePlanet'].value_counts()
```

```python
# not in : 포함되어 있지 않는다.
a1= ['Earth', 'Mars']
# HomePlanet의 값이 a1 리스트에 있는 것과 다른 것들만 가져온다.
a2 = df1.query('not HomePlanet in @a1')
a2['HomePlanet'].value_counts()
```
