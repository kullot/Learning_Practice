### 시트가 별도로 있는 경우의 엑셀 읽기
- pd.read_excel(파일경로, sheet_name=시트명|None )
- sheet_name 값이 없는 경우에는 첫번째 시트만 데이타프레임으로 저장
- sheet_name 값이 None인 경우에는 데이타프레임 딕셔너리로 저장

```python
# 데이타 확인 : shoppingcenter.xlsx
# 2 개의 시트로 구성 - 2018년 상반기 업력현황, 2018년 하반기 업력현황
```

6) 구별을 기준으로 정렬한 후 '구별' 컬럼을 인덱스로 지정하고 csv 파일로 저장하여라.

```python
df = pd.read_excel('data/population_in_Seoul.xls', header=2, usecols='B,D,G,J,N')
df
```

```python
c_list = ['구별', '인구수', '한국인', '외국인', '고령자']
df.columns
```

```python
df.columns = c_list
df
```

```python
df.drop(index=[0], inplace=True)
df
```

```python
df.dropna(how='all', inplace=True)
df
```

```python
df['외국인비율'] = df['인구수']/df['외국인']
df['고령자비율'] = df['인구수']/df['고령자']
df
```

```python
df.sort_values(by=['인구수'], ascending=False, inplace=True)
df
```

```python
list = []
count = 0
for i in df['구별'].values:
    count += 1
    list.append(f'{count}위 - {i}')
    print(f'{count}위 - {i}')
```

```python
df.sort_values(by=['구별'], inplace=True)
df
```

```python
df.to_excel('output\population_in_Seoul_sort.xlsx', encoding='cp949', index=False)
```

```python
# 6) 구별을 기준으로 정렬한 후 '구별' 컬럼을 인덱스로 지정하고 csv 파일로 저장하여라.
```
