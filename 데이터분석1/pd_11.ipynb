### sqlite Db 소스 확인

# SQLite
- 윈도우용 전용 프로그램

- sqlite(*.db) 데이타베이스파일 > 테이블 > 필드 + 행
- 데이타베이스 열기
### SQL Reference
- https://www.w3schools.com/sql/default.asp
### SQL 문법 특징
- sql 이란? db 조작 명령
- 대소문자를 구분 없음
- 마지막에 세미콜론(;)

### SELECT 절 => Read
SELECT 필드명|* FROM 테이블명 (WHERE 조건절 LIMIT 숫자);

### PANDAS 지원 SQL
- read_sql_query() => sql 구문 실행 후 데이타프레임
- read_sql_table() => DB안의 특정 테이블을 데이타프레임을 저장
- to_sql() => 데이타프레임을 DB안의 테이블로 저장

### sqlalchemy를 이용해서 DB 연결
- https://docs.sqlalchemy.org/en/14/core/engines.html#sqlite
```
# where <path> is relative:
engine = create_engine('sqlite:///foo.db')
```

````python
# 1) db/text1.db 데이타베이스랑 연결
url = 'sqlite:///db\\test1.db'
engine = create_engine(url)
engine
````

````python
# 2) db/test1.db > albums 테이블 => 데이타프레임으로 저장
# pd.read_sql_query(sql명령, engine)
sql = 'SELECT * FROM albums;'
df = pd.read_sql_query(sql, engine)
df.shape
````

````python
# 5) test1.db 데이타베이스의 tracks,  genres, media_tipes 테이블을 모두 연결하여라 
# 각각의 Name 컬럼명은 'Track Name', 'Genre Name', 'MediaType Name' 으로 변경하여야 한다
````

````python
url = 'sqlite:///db\\test1.db'
engine = create_engine(url)
````

````python
df_tracks = pd.read_sql_table('tracks', engine)
df_tracks.head(2)
````

````python
df_genres = pd.read_sql_table('genres', engine)
df_genres.head(2)
````

````python
df_media_types = pd.read_sql_table('media_types', engine)
df_media_types.head(2)
````

````python
# df_tracks + df_genres => 공통 컬럼이자 키는? GenreId
df2 = pd.merge(df_tracks , df_genres , on='GenreId')
df2.shape, df2.columns
````

````python
# Name 컬럼을 이름 변경 
df2.rename(columns={'Name_x':'Track Name', 
                    'Name_y':'Genre Name', 
                    'Name':'MediaType Name'}, inplace=True)
````

````python
df_tracks.rename(columns={'Name':'Track Name'}, inplace=True)
df_genres.rename(columns={'Name':'Genre Name'}, inplace=True)
df_media_types.rename(columns={'Name':'MediaType Name'}, inplace=True)
````

````python
# Name 공통컬럼 이지만 ValueError 발생
# pd.merge(df_tracks , df_genres , df_media_types)
````

````python
df = pd.merge(df_tracks,df_genres)
````

````python
df = pd.merge(df,df_media_types)
df.sample(5)
````

````python
# 7) customers, invoice_items, invoices 테이블을 모두 하나의 데이타프레임 df3 으로 저장하여라 
# 이때 출력되는 컬럼은 아래를 참조한다
````

````python
url = 'sqlite:///db\\test1.db'
engine = create_engine(url)
````

````python
df_customers = pd.read_sql_table('customers', engine)
df_customers.head(2)
````

````python
df_invoice_items = pd.read_sql_table('invoice_items', engine)
df_invoice_items.head(2)
````

````python
df_invoices = pd.read_sql_table('invoices', engine)
df_invoices.head(2)
````

````python
df3 = pd.merge(df_customers,df_invoices)
df3.head(2)
````

````python
df3 = pd.merge(df3,df_invoice_items)
df3.head(2)
````

````python
df3['Customer Name'] = df3['FirstName']+' '+df3['LastName']
````

````python
df3.head(2)
````

````python
df3 = df3[['CustomerId','Customer Name','Phone','Fax','Email','InvoiceId','InvoiceDate','TrackId','UnitPrice','Quantity']]
````
