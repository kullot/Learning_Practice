### pd.date_range() 함수 알아보기
- 연속적인 날짜 데이터에 사용
- pd.date_range(초기날짜, periods=count, freq = 주기코드값)
- pd.date_range(초기날짜, 마지막날짜, freq = 주기코드값)
- 주기코드 : freq
- https://pandas.pydata.org/pandas-docs/stable/user_guide/timeseries.html
```
s: 초
T: 분
H: 시간
D: 일(day)
B: 주말이 아닌 평일
W: 주(일요일)
W-MON: 주(월요일)
M: 각 달(month)의 마지막 날
MS: 각 달의 첫날
BM: 주말이 아닌 평일 중에서 각 달의 마지막 날
BMS: 주말이 아닌 평일 중에서 각 달의 첫날
WOM-2THU: 각 달의 두번째 목요일
Q-JAN: 각 분기의 첫달의 마지막 날
Q-DEC: 각 분기의 마지막 달의 마지막 날
```

### dt 접근자
- 시계열 데이타에 적용. 년,월,일,요일 추출 가능
- dt.year, dt.month, dt.day, dt.weekday(0이 월요일)
- dt.date # YYYY-MM-DD(문자)
- dt.year # 연(4자리숫자)
- dt.month # 월(숫자)
- dt.month_name() # 월(문자)
- dt.day # 일(숫자)
- dt.time # HH:MM:SS(문자)
- dt.hour # 시(숫자)
- dt.minute # 분(숫자)
- dt.second # 초(숫자)

````python
date = pd.date_range('2020-01', periods=1000)
df = pd.DataFrame(np.random.randn(1000,3), index=date, columns=list('ABC'))
df
````

````python
# 시계열인덱스를 특정 컬럼값으로 추가 
df['D'] = df.index
df.head()
````

````python
# D 컬럼에서 년도만 추출
df['D'].dt.year
````

````python
df['year'] = df['D'].dt.year
df['month'] = df['D'].dt.month
df['day'] = df['D'].dt.day
````

````python
df.sample(5)
````

````python
# 요일 컬럼 추가
df['weekDay'] = df['D'].dt.weekday
df.sample(5)
````

````python
# 0=>월요일 ~ 6=>일요일
df['weekDay'].unique()
````

````python
# weekDay 값 변경
df['weekDay2'] = df['weekDay'].replace({0:'월', 1:'화', 2:'수', 3:'목', 4:'금', 5:'토', 6:'일'})
````

````python
df.sample(5)
````

### 문자열(날짜데이타형식 , 년월일) => 날짜데이타

- pd.to_datetime(시리즈변수/데이타프레임[컬럼명], format='%Y%m%d', errors='coerce')
- errors='coerce' : 날짜데이타로 변경이 불가능한 경우 강제로 변환 NaN

````python
# 테스트용 문자열 형태으 ㅣ날짜 데이타 생성
# '년월일' => '20220629'
data_str = np.arange(20220101,20220111).astype(np.str)
data_str
````

````python
# 데이타프레임으로 변경
df = pd.DataFrame(data_str, columns=['Date1'])
df
````

````python
# 문자열 => 시계열 데이타
df['Date2'] = pd.to_datetime(df['Date1'], format='%Y%m%d', errors='coerce')
df.sample(3), df.info()
````

````python
# 년월일요일로 분리
df['Year'] = df['Date2'].dt.year
df['Month'] = df['Date2'].dt.month
df['Day'] = df['Date2'].dt.day
df['WeekDay'] = df['Date2'].dt.weekday
````

````python
df.sample(3)
````
