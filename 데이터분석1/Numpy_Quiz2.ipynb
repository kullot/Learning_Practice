```python
# 중복 확인
first_name = np.array(['박', '김', '이', '최', '신', '장', '윤'])
last_name1 = np.array(['은', '강', '종', '영', '준', '민'])
last_name2 = np.array(['철','수', '미', '원', '주', '혁', '희', '수'])

name_list = np.array([], dtype='<U3')

# 100 개 정도 생성하면 중복값이 나타난다 
for i in range(100):
    x = np.random.choice(first_name)
    y = np.random.choice(last_name1)
    z = np.random.choice(last_name2)
    temp = x + y + z
    name_list = np.insert(name_list, len(name_list), temp)
        
len(name_list), np.sort(name_list)
```

```python
# 중복이 없는 형태로 변경

# 1) 넘파이 배열 생성
first_name = np.array(['박', '김', '이', '최', '신', '장', '윤'])
last_name1 = np.array(['은', '강', '종', '영', '준', '민'])
last_name2 = np.array(['철','수', '미', '원', '주', '혁', '희', '수'])

# 2) 문자열 형태의 넘파이 배열 생성
name_list = np.array([], dtype='<U3')

# 3) 중복이 없다면 추가, 넘파이 길이가 7이면 while문 탈출
while True:
    x = np.random.choice(first_name)
    y = np.random.choice(last_name1)
    z = np.random.choice(last_name2)
    temp = x + y + z
    if len(name_list) >= 7:
        break
    if temp not in name_list:
        name_list = np.insert(name_list, len(name_list), temp)
        
# 4) 출력
len(name_list), np.sort(name_list)
```

```python
# 2) 1)의 넘파이 배열의 데이타 형태를 실수 형태로 변경하여라 
#   데이타의 결측치나 이상값은 -99로 지정한다. 
```

```python
# 1) 리스트 형태로 전처리
result = []
for row in data:
    temp = []
    try:
        # 날짜 - 삭제
        if row[0]:
            date = float(row[0].replace('-','').replace('-',''))
        else:
            date=-99
        temp.append(date)
        # 나머지 온도 데이타
        for item in row[1:]:
            
            try:
                temp.append(float(item))
            except:
                temp.append(-99)

#             if item == '':
#                 temp.append(-99)
#             else:
#                 temp.append(float(item))
    except:
        pass
    else:
        result.append(temp) 
```
