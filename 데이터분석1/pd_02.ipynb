#### 데이타프레임 생성방법
- 2차원 리스트 | 딕셔너리 리스트 | 딕셔너리 내포 | 시리즈 | 넘파이배열 => 데이타프레임
- csv, tsv, dat, excel, txt 외부파일 => 데이타프레임
- db => 데이타프레임
- web상의 table 구조 => 데이타프레임
- open API (주식, 환율, 원자재 ...) => 데이타프레임

##### 데이타프레임이란?
- 데이타프레임은 시리즈의 조합이다.
- 행과 열로 구성 => 엑셀의 테이블과 비슷한 스타일

##### DataFrame 생성하기 1
- 딕셔너리 리스트 => 데이타프레임
- 딕셔너리리스트 구조 : 딕셔너리 키 하나당 리스트1개

##### {키1:[값1, 값2 ], 키2:[값1, 값2 ]...}
- 데이타프레임변수 = pd.DataFrame(딕셔너리리스트)
- 행의 인덱스가 숫자 / 컬럼인덱스는 딕셔너리의 키값

#### 2차원리스트 => 데이타프레임
```
데이타프레임변수 = pd.DataFrame(2차원리스트,
                         index = 인덱스리스트,
                         columns = 컬럼리스트)
```

````python
# 주소록 리스트
myList = [['홍길동', '부산', '010-3333-8888'],
         ['최길동', '울산', '010-5673-8888'],
         ['박길동', '마산', '010-2222-8888']]
myList
````

````python
# 데이타프레임화
df = pd.DataFrame(myList, columns = ['name', 'addr', 'mobile'], index=['num1', 'num2', 'num3'])
df
````

### Dataframe : 행과 열 조회

### 특정컬럼
- 데이타프레임이름[컬럼명]
- 데이타프레임이름[[컬럼명1, 컬럼명2 ... ]]

### 특정행의 특정컬럼
- 데이타프레임이름.iloc[행인덱스순서] : 시리즈형태로 세로로 추출
- 데이타프레임이름. iloc[ row1:row2, column1:column2]
- 데이타프레임이름.iloc[ [인덱스 숫자], [컬럼인덱스 숫자 ]]

### 특정 행¶
- 데이타프레임명.loc[행이름, :] => 시리즈
- 데이타프레임명.loc[행이름] => 시리즈
- 데이타프레임명.iloc[행인덱스, :] => 시리즈
- 데이타프레임명.iloc[행인덱스] => 시리즈

### 다중행
- 데이타프레임명.loc[startRowLabel:endRowLabel, :]
- 데이타프레임명.iloc[startRowIndex:endRowIndex, :]

````python
df = pd.DataFrame( np.arange(1,26).reshape(5,5), 
                   index=[ 'row'+str(i) for i in range(1,6)], 
                   columns=[ 'column'+str(i) for i in range(1,6) ])

print(type(df), df.shape, df.size)
df
````

````python
# 특정컬럼
# 데이타프레임이름[컬럼명], 데이타프레임이름.컬럼명 => 시리즈
# 데이타프레임이름[[컬럼명1, 컬럼명2 ... ]] => 데이타프레임

# 하나의 컬럼
print(df.column1)
print()
print(df['column1'])
print()

# 다중컬럼
print(df[['column1', 'column3']])
df[['column1', 'column3']]
````

````python
# 하나의 행 => 시리즈
print(df.loc['row1'])
print()
print(df.iloc[0])
````

````python
# 하나의 행 => 데이타프레임
# 데이타프레임명.loc[[행이름]]
# 데이타프레임명.iloc[[행인덱스]]
print(df.loc[['row1']])
print()
df.iloc[[0]]
````

````python
# 다중행 => 데이타프레임
# 데이타프레임명.loc[[행이름1, 행이름2, ...]]
# 데이타프레임명.iloc[[행인덱스1, 행인덱스2, ...]]
print(df.loc[['row1','row3','row5']])
print()
df.iloc[[0,2,4]]
````

````python
# 다중행 => 데이타프레임 (연속행만 가능, 슬라이싱)
# 데이타프레임명.loc[행이름1:행이름2]
# 데이타프레임명.iloc[행인덱스1:행인덱스2]
print(df.loc['row2':'row4'])
print()
df.iloc[1:4]
````

````python
# 컬럼 인덱싱 + loc, iloc 사용
# 다중컬럼 => 데이타프레임 (연속컬럼만 가능, 슬라이싱)
# 데이타프레임명.loc[:, 컬럼이름1:컬럼이름2]
# 데이타프레임명.iloc[:, 컬럼인덱스1:컬럼인덱스2]
df
````

````python
print(df.loc[:, 'column2':'column4'])
df.iloc[:, 1:4]
````
