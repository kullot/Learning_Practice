### concat() 메소드 이용하기
- 세로로 합치기 : 행방향
    - 새로운데이타프레임 = pd.concat([데이타프레임1, 데이타프레임2 .. ])
- 가로로 합치기 : 열방향
    - 새로운데이타프레임 = pd.concat([데이타프레임1, 데이타프레임2 .. ], axis=1, ignore_index=True)
    - 적용 데이타프레임의 행수가 다른 경우 누락값(NaN) 발생
- 중복 컬럼값이 없게 합치기 : ingnore_index=True

````python
# 테스트용 데이타프레임 생성
# 조건 : 컬럼명이 같아야한다.
df1 = pd.DataFrame(np.arange(1,17).reshape(4,4), columns=list('ABCD'))
df1
````

````python
df2 = pd.DataFrame(np.arange(21,37).reshape(4,4), columns=list('ABCD'))
df2
````

````python
df3 = pd.DataFrame(np.arange(41,57).reshape(4,4), columns=list('ABCD'))
df3
````

### concat() + 이너조인(Inner Join) 옵션 이용하기

- 조인(join) : 데이타프레임 연결 방식
    - 내부조인(=이너조인 inner join) : 공통 컬럼 또는 공통 인덱스
    - 외부조인(=아우터조인 outer join)

- 공통컬럼만 골라 연결하기
outer 가 기본값 join='outer'

pd.concat([데이타프레임1, 데이타프레임2..], ignore_index=True, 
            join='inner')


- 공통행만 골라 연결하기

pd.concat([데이타프레임1, 데이타프레임2..], axis=1, 
            join='inner')

### merge() 메소드1
- 두 데이터 프레임의 공통 컬럼이나 인덱스를 기준으로 두 개의 테이블을 합친다.
- 이 때 기준이 되는 컬럼과 인덱스를 키(key)라고 한다.
- SQL 스타일
    - pd.merge(데이타프레임1, 데이타프레임2) : Inner Join
    - pd.merge(데이타프레임1, 데이타프레임2, how='outer/left/right')

````python
# 샘플 데이타프레임 생성 
# 공통컬럼 : 고객번호 

df1 = pd.DataFrame({
    '고객번호': [1001, 1002, 1003, 1004, 1005, 1006, 1007],
    '이름': ['둘리', '도우너', '또치', '길동', '희동', '마이콜', '영희']
}, columns=['고객번호', '이름'])

df2 = pd.DataFrame({
    '고객번호': [1001, 1001, 1005, 1006, 1008, 1001],
    '금액': [10000, 20000, 15000, 5000, 100000, 30000]
}, columns=['고객번호', '금액'])
````

````python
df1
````

````python
df2
````

````python
# 공통컬럼의 공통키값으로 연결된 데이타만 출력1
df3 = pd.merge(df1, df2) 
df3
````

````python
# 공통컬럼의 공통키값으로 연결된 데이타만 출력2
df4 = pd.merge(df2, df1) 
df4
````

````python
# 아우터 이너조인1 - 모든 데이타프레임의 컬럼 값이 함께 출력  
# pd.merge(데이타프레임1, 데이타프레임2, how='outer/left/right')
df5 = pd.merge(df1, df2, how='outer') 
df5
````

````python
# 아우터 이너조인2 - 모든 데이타프레임의 컬럼 값이 함께 출력  
# pd.merge(데이타프레임1, 데이타프레임2, how='outer/left/right')
# df1 데이타프레임의 모든 값이 출력된다
df6 = pd.merge(df1, df2, how='left') 
df6
````

````python
# 아우터 이너조인3 - 모든 데이타프레임의 컬럼 값이 함께 출력  
# pd.merge(데이타프레임1, 데이타프레임2, how='outer/left/right')
# df2 데이타프레임의 모든 값이 출력된다 
# df1 데이타프레임에서 df2의 고객번호가 아닌것은 데이타가 누락된다. => 1002, 1003, 1004, 1007 번 누락 
df7 = pd.merge(df1, df2, how='right') 
df7
````

### merge() 메소드2

- 기준 컬럼명 지정하여 데이타병합하기
- on = '기준컬럼명'
    - 이름이 같은 컬럼이 있다면 _x 또는 _y 와 같은 접미사가 붙는다.

````python
df1 = pd.DataFrame({
    '고객명': ['춘향', '춘향', '몽룡'],
    '날짜': ['2018-01-01', '2018-01-02', '2018-01-01'],
    '데이터': ['20000', '30000', '100000']})
df1
````

````python
# df1과 공통 컬럼 고객명, 데이터 컬럼이 존재한다
# 데이터 컬럼의 경우 데이터형이 숫자글자, 글자 형태이다
df2 = pd.DataFrame({
    '고객명': ['춘향', '몽룡'],
    '데이터': ['여자', '남자']})
df2
````

````python
# 이너조인 테스트 1
df_a = pd.merge(df1, df2)
df_a
````

````python
# 이너조인 테스트 2
df_b = pd.merge(df1, df2, how='outer')
df_b
````

````python
# 이너조인 테스트 3 - 고객명 컬럼만 공통키로 설정
df_c = pd.merge(df1, df2, on='고객명')
df_c
````

### merge() 메소드2
- left_on== '기준컬럼명1', right_on== '기준컬럼명2'
    - 키가 되는 기준열의 이름이 두 데이터프레임에서 서로 다른 경우 사용.
- 두 데이타프레임에서 공통 컬럼명이 있지만 데이터형이 다른 경우
- on = '기준이되는컬럼명' 지정

````python
dfSt1 = pd.DataFrame({
    '이름': ['영희', '철수', '철수'],
    '성적A': [1, 2, 3]})
dfSt1
````

````python
# dfSt1과 비교해서 공통컬럼명이 다르지만 성명이 dfSt1의 이름 컬럼에 해당한다 
dfSt2 = pd.DataFrame({
    '성명': ['영희', '영희', '철수'],
    '성적B': [4, 5, 6]})
dfSt2
````
