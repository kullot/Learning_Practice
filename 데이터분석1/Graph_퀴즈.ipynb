### 5) 4번의 folium 지도에서 서울시 구별,  진료소의 갯수를 기반으로  choropleth와 연동하여 지도를 완성하여라 

```python
# 2) json 데이타 => 변수에 저장
json_path = open('data/skorea_municipalities_geo_simple.json', 'r', encoding='utf-8')
json_str = json.load(json_path)

type(json_path), type(json_str)
```

```python
df2 = df.groupby('시군구').count()
df2.sample(5)
```

```python
# 폴리움 객체 생성 
map = folium.Map( location=[37.5502, 126.982], 
                  zoom_start=11,
                 # tiles='Stamen Watercolor',
                )
map.choropleth(geo_data=json_str, data=df2,
               columns=[df2.index, '주소'],
               key_on='feature.id',
               fill_color='YlGn', # Purples
               line_color='black',
               line_weight=1,
               line_opacity=0.5,
               legend_name='서울시 구별 진료소 갯수 (%)')
# 마커 표시
for i in range(len(df)):
    x = df['위도'][i]
    y = df['경도'][i]
    msg = df.index[i]
    folium.Marker([x,y],
                  popup=msg,
                  icon=folium.Icon(icon='building', color='blue', prefix='fa')).add_to(map)
# 폴리움 호출
map
```
