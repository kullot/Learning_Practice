````python
# 1) 위도+경도 데이타 구하기 
# 예시) 광안역 
# - https://xn--yq5bk9r.com/blog/map-coordinates
# 광안역좌표 : 위도 : 35.15791190000015, 경도 : 129.10896279999997
# 수영역좌표 : 위도 : 35.16547460000029, 경도 : 129.11057529999997
# 광안1동 행정복지센터 
# 좌표 : 위도 : 35.162893000000274, 경도 : 129.10821069999932
````

### 2) 폴리움을 이용한 마커 표시
```
지도객체 = folium.Map( location=[위도, 경도], zoom_start=확대배율숫자
                       , tiles=타일스타일)
folium.Marker([위도2,경도2], popup='마커메세지', 
       icon=folium.Icon(icon='아이콘코드값', color='색상코드값',      
                        prefix='fa')).add_to(지도객체)
```

## 아이템 코드값 확인은?
- 타일스타일
    - "OpenStreetMap"
    - "Mapbox Bright" (Limited levels of zoom for free tiles)
    - "Mapbox Control Room" (Limited levels of zoom for free tiles)
    - "Stamen" (Terrain, Toner, and Watercolor)
    - "Cloudmade" (Must pass API key)
    - "Mapbox" (Must pass API key)
    - "CartoDB" (positron and dark_matter)
- 지도객체를 html 파일로 저장하기
    - 지도객체.save('경로/파일명.html')

### folium.CircleMaker
- 반경과 색상 지정
```
folium.CircleMarker([위도, 경도], radius=반지름, 
                    popup='팝업메세지', color='색상값', 
                    fill_color='색상값', ).add_to(foliumMap객체)
```

````python
# 폴리움 객체 생성 
map = folium.Map(location=[35.15791190000015, 129.10896279999997], 
                 zoom_start=15,
                 tiles='Stamen Watercolor')
# 마커 표시
folium.Marker([35.1579124,129.113145],
              
              popup='광안역',
              icon=folium.Icon(icon='subway', color='green', prefix='fa')).add_to(map)

folium.Marker([35.16623791,129.11578896],
              popup='수영역',
              icon=folium.Icon(icon='subway', color='green', prefix='fa')).add_to(map)

folium.Marker([35.1628628,129.1123608],
              popup='광안1동 행정복지센터',
              icon=folium.Icon(icon='coffee', color='red', prefix='fa')).add_to(map)

folium.CircleMarker([35.1510975,129.1116415], 
                    radius=50, 
                    fill_color='blue', ).add_to(map)

# html 문서로 저장
map.save('output/map1.html')

# 폴리움 호출
map
````
