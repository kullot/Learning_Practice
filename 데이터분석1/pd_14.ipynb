### 그룹화 - groupby()
- 데이타프레임.groupby(컬럼명).컬럼명.집계함수
- 그룹객체명 = 데이타프레임.groupby(컬럼명)
- 그룹객체명.집계함수()
- 그룹객체명.agg([집계함수1, 집계함수2...])
- 집계함수의 종류
    - count, size, mean, std, min, max, sum, var

```python
# 대륙별 평균수명(lifeExp)의 평균
# 데이타프레임.groupby(그룹핑컬럼명).수치형컬럼명.집계함수 => 시리즈 
print(df['year'].unique())
df.groupby('continent').lifeExp.mean()
```

```python
# 대륙별 인구의 최대값 
# 데이타프레임.groupby(그룹핑컬럼명).수치형컬럼명.집계함수 => 시리즈 
print(df['continent'].unique())
print()
df.groupby('continent').pop.max()
```

```python
df['continent'].value_counts()
```

```python
# 대륙별 각 년도의 데이터수 출력 
# 데이타프레임.groupby(그룹핑컬럼명).수치형컬럼명.집계함수 => 시리즈 
print(df['year'].unique())
df.groupby('continent').year.count()
```

```python
# 그룹 객체화 + 집계함수 
# 1) 그룹객체
#    그룹객체명 = 데이타프레임.groupby(컬럼명)
# 2) 집계함수 적용 
#    그룹객체명.집계함수()
```

```python
# 1) 그룹객체 생성 - 대륙별
#    그룹객체명 = 데이타프레임.groupby(컬럼명)
group_c = df.groupby('continent')
group_c
```

```python
# 그룹객체명.groups => (대륙별이 키)
print(type(group_c.groups))
group_c.groups
```

```python
# 키로 접근 => 키값에 해당하는 인덱스
group_c.groups['Africa']
```

```python
df[df['continent'] == 'Africa'].index
```

```python
# 2) 집계함수 적용 
#    그룹객체명.집계함수() => 데이타프레임
# 모든 수치데이타의 컬럼의 평균 데이타프레임
group_c.mean()
```

```python
df.groupby('continent').year.mean()
```

```python
# 대륙별 lifeExp, pop 컬럼의 평균만 출력
df.groupby('continent').mean()[['lifeExp','pop']]
```

```python
# 년도별 lifeExp, pop, gdpPercap 컬럼의 평균만 출력
df.groupby('year').max()[['lifeExp','pop','gdpPercap']]
```

```python
# 1952년도 평균수명의 최대값 
# 72.670
df.groupby('year').max()[['lifeExp', 'pop', 'gdpPercap' ]].loc[1952, 'lifeExp']
```

```python
df[df['year'] == 1952]['lifeExp'].max()
```

### 그룹핑 + 다중 집계함수
- 데이타프레임명.groupby(그룹기준의컬럼).agg([집계함수1, 집계함수2 ...])
- 그룹객체명.agg([집계함수1, 집계함수2 ...])
- 집계함수란? sum(), mean(), max(), min(), std(), var(), count()....
- 집계함수 리스트는 넘파이집계함수명을 데이타로 한다 

```python
df.groupby('continent').agg([np.sum, np.mean])
```

```python
# 대육별 평균수명과 인구의 총합과 평균
df.groupby('continent')[['lifeExp', 'pop']].agg([np.sum, np.mean])
```

```python
# 대륙별 평균수명의 최대값과 최소값 
df.groupby('continent')['lifeExp'].agg([np.min, np.max])
```
