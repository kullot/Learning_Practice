### 리스트 데이타 => 그래프

````python
# 1) 데이타 생성 
x_list = list(range(-10, 11))
y_list = []
for item in x_list:
    y_list.append(item**2)
print(x_list, len(x_list))
print(y_list, len(y_list))
````

````python
# 2) 그래프화
# plt.plot(x, y)
plt.plot(x_list, y_list)
````

````python
plt.figure(figsize=(10,7))
plt.title('y = x**2 그래프')
plt.plot(x_list, y_list, color='red', linestyle='--', 
         marker='o', linewidth=2, markerfacecolor='black',
         label='graph1')
plt.legend()
plt.grid()

# 이미지 파일로 저장
plt.savefig('output/graph1.png')
plt.show()
````

````python
plt.figure(figsize=(10,7))
plt.title('y = x**2 그래프')
plt.plot(x_list, y_list, color='red', linestyle='--', 
         marker='o', linewidth=2, markerfacecolor='black',
         label='graph1')
plt.legend()
plt.grid()

# 축 제목
plt.xlabel('X 축') 
plt.ylabel('Y 축') 

# 축의 범위 설정
plt.xlim(-10, 10)
plt.ylim(0, 100)

# 눈금 설정
plt.xticks(np.arange(-10, 11)) 
plt.yticks(np.arange(0, 101, 10)) 

# 이미지 파일로 저장
plt.savefig('output/graph2.png')

plt.show()
````

### 레이아웃 배치 기법1

- 위치 지정
    - plt.subplot(행 열 순서)
    
- 그래프 
    - 데이타프레임[인덱스명].plot(kind=그래프종류, title=제목)

````python
# 샘플 데이타프레임생성 
data = {'kor':[40, 55, 66, 77, 34],
        'math':[70, 55, 66, 20, 40],
        'eng':[80, 95, 100, 77, 90],
        'sci':[90, 45, 77, 90, 67]}
        
name_list = ['고길동','김희동','신윤지','이민호','박준희']    

df = pd.DataFrame(data, index=name_list)
df
````

````python
# 2) 2행 2열로 배치 

# 그래프 사이즈 지정 
plt.figure(figsize=(12, 10))

# 2행 2열의 첫번째 
# plt.subplot(행 열 순서)
plt.subplot(221)
df['kor'].plot( kind='bar', title='kor')

plt.subplot(222)
df['math'].plot( kind='bar', title='math')

plt.subplot(223)
df['eng'].plot( kind='bar', title='eng')

plt.subplot(224)
df['sci'].plot( kind='bar', title='sci')

# 여백 지정 
plt.tight_layout()
````

````python
# 3) for문 + 2행 2열

# 그래프 사이즈 지정 
plt.figure(figsize=(12, 10))
c_list = list(df.columns)

for i in range(4):
    plt.subplot(221+i)
    df[c_list[i]].plot( kind='bar', title=c_list[i])
    
# 전체 제목 설정
plt.suptitle('중간 고사 성적 그래프')

# 여백 지정 
plt.tight_layout()

plt.show()
````

### 파이 그래프
```
plt.pie(데이타프레임,  
        autopct='%전체자릿수.소숫점이하자릿수f%%', 
        colors=[파이조각색상, ...],
        labels = [라벨명1, ...],
        startangle=각도,
        explode=[거리값 ...])
```

````python
# 1) 데이타프레임 생성
df = pd.read_csv('data/wages.csv')
df.head()
````

````python
# 2) sex 컬럼의 분포값
df['sex'].value_counts()
````

````python
df['sex'].value_counts().index
````

````python
# 3) 그래프화 
# 색상 조합 
# https://colorhunt.co/palettes/popular
plt.figure(figsize=(10,8))
plt.pie(df['sex'].value_counts(),  
        autopct='%3.2f%%', 
        colors=['red', 'blue'],
        labels = df['sex'].value_counts().index,
        startangle=-15,
        explode=[0,0.05]
       )
plt.legend()
plt.axis('equal')
plt.show()
````

````python
# race 적용
df['race'].value_counts()
````

````python
plt.figure(figsize=(10,8))
plt.pie(df['race'].value_counts(),  
        autopct='%3.2f%%', 
        colors=['#FCF8E8', '#DF7861', '#ECB390', '#D4E2D4'],
        labels = df['race'].value_counts().index,
        startangle=-15,
        # explode=[0,0.05]
        )
plt.legend()
plt.axis('equal')
plt.show()
````
