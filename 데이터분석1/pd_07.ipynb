### 퀴즈
- 1) url의 json을 데이타프레임 형태로 저장하여라
    - url = http://jsonplaceholder.typicode.com/users

- 2) 모두 몇명인가?

- 3) 'id', 'name', 'username', 'email', 'phone', 'website' 컬럼만 별도의 데이타프레임으로 저장하여라

```python
url = 'http://jsonplaceholder.typicode.com/users'
df = pd.read_json(url)
df.sample(5)
```

```python
print(df.shape[0], '명')
```

```python
df2 = df[['id', 'name', 'username', 'email', 'phone', 'website']].copy()
df2
```

```python
df.loc[0,'address']
```

```python
df['address'][0]
```

```python
# KeyError
# df['address']['street']
df['address'].str['street']
```

```python
df2['street'] = df['address'].str['street']
df2['suite'] = df['address'].str['suite']
df2['city'] = df['address'].str['city']
df2['zipcode'] = df['address'].str['zipcode']
df2
```

```python
df2['lat'] = df['address'].str['geo'].str['lat']
df2['lng'] = df['address'].str['geo'].str['lng']
df2
```

```python
df['company'][0]
```

```python
df2['company_name'] = df['company'].str['name']
df2['company_catchPhrase'] = df['company'].str['catchPhrase']
df2['company_bs'] = df['company'].str['bs']
df2
```

```python
df2.to_csv('output/user.csv', index= False)
```
