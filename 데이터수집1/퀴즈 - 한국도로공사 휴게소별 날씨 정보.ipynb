```python
# 1) 고속도로 휴게소별 날씨 정보 Api 서비스를 신청하여라 
url = 'http://data.ex.co.kr/openapi/restinfo/restWeatherList?key=9408629118&type=json&sdate=20220602&stdHour=10'
response = requests.get(url)
print(response)
# print(response.text)
```

```python
# 2) 신청한 api 서비스를 이용하여 특정 날짜(예시-2022년6월2일)의 날씨 데이타를 
# 리스트 딕셔너리 형태로 저장하여라. 
# 아래 예시는 첫번째 고속도로 휴게소의 데이타이다. 
weather = json.loads(response.content)
# weather
weather_list = weather['list']
print(type(weather_list))
weather_list[0]
```

```python
# 4) 경부선에 위치한 휴게소의 날씨 데이터만 아래와 같이 출력하여라 
for i in weather_list:
    if i['routeName'] == '경부선':
        print(i['routeName'], '-', i['unitName'], i['weatherContents'], i['tempValue'][:4],"도" )
```

```python
# 5) 데이프레임으로 저장하고 아래와 같이 출력하여라 
df = pd.DataFrame(weather_list)
df
```

```python
# 6) 2022년 5월 29~31일 3일간의 데이타를 API 서비스를 이용하여 추출하고 
# 데이타프레임으로 저장하여라
result_list = []
for number in range(29, 32):
    url = f'http://data.ex.co.kr/openapi/restinfo/restWeatherList?key=5464311301&type=json&sdate=202205{number}&stdHour=10'
    print(url)
    response = requests.get(url)
    print(response)
#     print(response.text)
    weather = json.loads(response.content)
    weather_list = weather['list']
    for item in weather_list:
        result_list.append(item)
df = pd.DataFrame(result_list)
df
```
