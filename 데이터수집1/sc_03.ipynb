## 블로터의 실시간랭킹 텍스트만 출력하여라 
https://www.bloter.net/

```python
# (1) 웹페이지 접근
# 샘플 url => html 소스 => 문자열로 변환
# 예제 소스 : 블로터 : headers 정보가 없으면 304 코드. 페이지 내용이 없다

# 객체화
url = 'https://www.bloter.net'
res = requests.get(url)
res, res.text
```

```python
# 304 상태코드 => 200 상태코드

# 사이트 이용 => http://www.useragentstring.com/

# 유저정보 생성
user_agent_info = 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/101.0.4951.64 Safari/537.36 Edg/101.0.1210.53'
# 해더 정보 생성
headers = {'User-Agent' : user_agent_info}
# 객체화
url = 'https://www.bloter.net'
res = requests.get(url, headers=headers)
# res, res.text

# 객체로부터 html 페이지 생성
html_str = res.text
# html_str
```

```python
# 2) html 소스 => BeautifulSoup 객체화
# soup변수명 = BeautifulSoup(html변수, 해석기(html.parser/html5/lxml/xml))
soup = BeautifulSoup(html_str, 'html.parser')
print(type(soup))
# soup
```

## 영문 색상명 + 16진수색상값
- 예제 사이트 : https://www.w3schools.com/colors/colors_names.asp
- 영문색상명, 16진수 색상값, URL정보 => color.csv

```python
# 1) 웹페이지 접근 => response 객체 => html 소스
# 예제 소스 : w3 schools > html color name

# 객체화
url = 'https://www.w3schools.com/colors/colors_names.asp'
res = requests.get(url)
print(res)
html_str = res.text
# html_str
```

```python
# 2) html 소스 => BeautifulSoup 객체화
# soup변수명 = BeautifulSoup(html소스, 해석기(html.parser/html5/lxml/xml))
soup = BeautifulSoup(html_str, 'html.parser')
print(type(soup))
soup
```

```python
# 3) 소스 파악
# 
```

```python
# 4) 색상 리스트 생성
color_list = soup.select('.innerbox')
# color_list = soup.find_all(class_='innerbox')
print(len(color_list))
print('='*50)
print(color_list[0])
print('='*50)
print(color_list[-1])
print('='*50)
```

```python
# 5) 첫번째 색상 정보만 추출
# 색상명
print('색상명 :', color_list[0].find('a').text)
# a 태그중에서 2변째의 텍스트
print('16진수 색상값 :', color_list[0].find_all('a')[1].text)
# 색상 URL
base_url = 'http://www.w3schools.com'
print('색상 URL :', base_url+color_list[0].find('a')['href'])
print('='*50)
print(color_list[0])
```

```python
# 6) 2차원 리스트로 저장
# 빈리스트 생성
color_list2 = []
base_url = 'http://www.w3schools.com'
for color in color_list:
    print('색상명 :', color.find('a').text)
    print('16진수 색상값 :', color.find_all('a')[1].text)
    print('색상 URL :', base_url+color.find('a')['href'])
    # 리스트에 저장
    color_list2.append([color.find('a').text, color.find_all('a')[1].text, base_url+color.find('a')['href']])
print('='*50)
print(len(color_list2))
```

```python
# 2차원 리스트를 출력
for i, j, k in color_list2:
    print(i, j, k)
```

```python
# 7) 2차원 리스트 => 판다스의 데이타프레임 구조
# 판다스 임포트
import pandas as pd

pd.__version__
```

```python
# 데이타프레임명 = pd.DataFrame(2차원리스트, columns=[컬럼명 ...])

df = pd.DataFrame(color_list2, columns=['색상명', '16진수 색상값', 'URL'])
df
```

```python
# 8)데이타프레임 => csv 파일로 저장
# 데이타프레임명.to_csv('csv저장파일경로', index=False, encoding='cp949,utf-8')
df.to_csv('output/color.csv', index=False, encoding='cp949')
```
