### 퀴즈
- 1) 동네예보 > 중기예보 > 전국 에서 도시명(city 태그 이용)만 출력하여라

- [중기예보]-[전국]
http://www.kma.go.kr/weather/forecast/mid-term-rss3.jsp?stnId=108

```python
url = 'https://www.kma.go.kr/weather/forecast/mid-term-rss3.jsp?stnId=108'
res = requests.get(url)
xml_str = res.text
# xml_str
```

```python
soup = BeautifulSoup(xml_str, 'xml')
soup
```

```python
target_list = soup.find_all('location')
# target_list[0]
len(target_list)
```

```python
# 4) 도시만 출력 
target_city = soup.find_all('city')
target_city
```

```python
city_list = []
for city in target_city:
    city_list.append(city.text)
print(city_list)
```

```python
# 리스트 포 = 리스트 컴프리핸션
# 리스트명 = [ 결과값 for 문 ]
city_list2 = [city.text for city in target_city]
print(city_list2)
```

### 2) '동네예보 > 중기예보 > 전국' 의 날씨 데이타를 아래와 같은 형태로 데이타프레임으로 변경하여라

```python
target_list = soup.find_all('location')
midterm_forecast_list2 = []

for location in target_list:
    city = location.city.text
    
    target_data_list = location.find_all('data')
    for data in target_data_list:   
        tmEf = data.tmEf.text
        wf = data.wf.text
        tmn = data.tmn.text
        tmx = data.tmx.text
        rnSt = data.rnSt.text
        
        midterm_forecast_list2.append([city, tmEf, wf, tmn, tmx, rnSt])
```

```python
df = pd.DataFrame(midterm_forecast_list2, columns=['도시', '날짜', '날씨', '최저기온', '최고기온', '강수률'])
df
```

### 3) '동네예보 > 중기예보 > 전국' 에서 부산, 대구, 원주의 날씨 데이타만 아래와 같은 형태로 데이타프레임으로 변경하여라

```python
target_list = soup.find_all('location')
midterm_forecast_list3 = []

for location in target_list:
    city = location.city.text
    
    if city in ['부산', '대구', '원주']:
        target_data_list = location.find_all('data')
        for data in target_data_list:   
            tmEf = data.tmEf.text
            wf = data.wf.text
            tmn = data.tmn.text
            tmx = data.tmx.text
            rnSt = data.rnSt.text
        
            midterm_forecast_list3.append([city, tmEf, wf, tmn, tmx, rnSt])
```

```python
df = pd.DataFrame(midterm_forecast_list3, columns=['도시', '날짜', '날씨', '최저기온', '최고기온', '강수률'])
df
```
