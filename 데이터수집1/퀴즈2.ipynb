```python
# 관련 모듈 임포트 
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = 'https://comic.naver.com/webtoon/weekday'
res = requests.get(url)
html_str = res.text
soup = BeautifulSoup(html_str)
soup
```

```python
target1 = soup.select('div.col>div>ul>li')
target2 = soup.select('div.col>div>h4')
# target1
# target2
```

```python
base_url = 'https://comic.naver.com'
month = target1[0].select_one('a')['href']
title = target1[0].select_one('img')['alt']
URL = base_url + target1[0].select_one('a')['href']
print('요일 :', month[-3:])
print('제목 :', title)
print('URL :', URL)
# target[0]
```

```python
# 6) 2차원 리스트 구조로 변경
# 빈리스트 생성
comic_list = []
base_url = 'https://comic.naver.com'
for li in target1:
    month = li.select_one('a')['href']
    title = li.select_one('img')['alt']
    URL = base_url + li.select_one('a')['href']
    print(month[-3:], title, URL)
    # 리스트에 추가
    comic_list.append([month[-3:], title, URL])
```

```python
# 데이타프레임 변경
df = pd.DataFrame(comic_list, columns = ['month', 'title', 'URL'])
df
```
