```python
# 0) 모듈 임포트
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

## Yes24 베스트셀러 정보 저장하기
- http://www.yes24.com/Main/default.aspx
- 1~10위 까지 베스트셀러(순위, 책제목, 제자, 출판사, 표지_URL, 책상세_URL)
- 문자열함수 문법(replace(), split(), 인덱싱, 슬라이싱)

```python
# 1) 웹페이지 접근 => reponse 객체 => html 페이지
url = 'http://www.yes24.com/Main/default.aspx'
res = requests.get(url)
html_str = res.text
print(res)
# html_str
```

```python
# 2) 뷰티풀숩 객체화
# html 페이지 => 숩객체
soup = BeautifulSoup(html_str)
soup
```

```python
# 4) 순위 정보 추출
# tp02 클래스값을 가지고 있는 li 태그
target = soup.select('li.tp02')
print(len(target))
# 1위만
target[0]
# 10위만
target[-1]
```

```python
# 5) 1위에서 개별 정보 추출
# (순위, 책제목, 제자, 출판사, 표지_URL, 책상세_URL)
print('순위 => ', target[0].select_one('strong').text)
# 2번재 strong 태그에서 텍스트만
print('책제목 => ', target[0].select('strong')[1].text)
# 2번재 em 태그에서 텍스트만
print('저자 => ', target[0].select('em')[1].text)
# a 태그안에 title 속성
print('출판사 => ', target[0].select_one('a')['title'])
# split()을 이용하여 공백을 기준으로 분리 => 리스트
print('출판사 => ', target[0].select_one('a')['title'].split())
# 마지막 인덱스가 출판사
print('출판사 => ', target[0].select_one('a')['title'].split()[-1])
# 작은 표지 이미지 경로(S) => 큰 표지 이미지 경로(XL)
print('표지 URL => ', target[0].select_one('img')['src'])
print('표지 URL => ', target[0].select_one('img')['src'].replace('S','XL'))
# http://www.yes24.com/Product/Goods/
base_url = 'http://www.yes24.com'
print('상세 URL => ', base_url + target[0].select_one('a')['href'])
print('='*50)
target[0]
```

```python
# 6) 2차원 리스트 구조로 변경
# 빈리스트 생성
bestbook_list = []
base_url = 'http://www.yes24.com'
for li in target:
    rank = li.select_one('strong').text
    title = li.select('strong')[1].text
    writer = li.select('em')[1].text
    pub = li.select_one('a')['title'].split()[-1]
    img_url = li.select_one('img')['src'].replace('S','XL')
    book_url = base_url+li.select_one('a')['href']
    print(rank, title, writer, pub, img_url, book_url)
    # 리스트에 추가 
    bestbook_list.append([rank, title, writer, pub, img_url, book_url])
```

```python
# 2차원 리스트에서 출력
for rank, title, writer, pub, img_url, book_url in bestbook_list:
    print(rank, title, writer, pub, img_url, book_url)
bestbook_list
```

```python
# 7) 2차원리스트 => 데이타프레임 구조로 변경
df = pd.DataFrame(bestbook_list, columns=['rank', 'title', 'writer', 'pub', 'img_url', 'book_url'])
df
```
