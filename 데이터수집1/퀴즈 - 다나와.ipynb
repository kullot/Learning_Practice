```python
import pandas as pd
import time
from selenium import webdriver
from bs4 import BeautifulSoup
import requests
```

```python
driver = webdriver.Chrome('chromedriver.exe')
time.sleep(2)
url = 'https://www.danawa.com'
driver.get(url)
```

```python
# search_word = input('검색어 입력하기 =>').strip()
driver.find_element_by_name('k1').send_keys('삼성전자 갤럭시 워치 액티브2')
driver.find_element_by_name('k1').submit()
```

```python
html_str = driver.page_source
soup = BeautifulSoup(html_str)
target = soup.select('div.prod_main_info')
len(target)
```

```python
# 제품명	링크	가격	평점	등록일	스펙

title = target[0].select_one('img')['alt']
URL = target[0].select_one('a')['href']
price = target[0].select_one('p.price_sect').text.strip().replace('\n','').replace('가격정보 더보기','')
if target[0].select_one('div.point_num'):
    score = target[0].select_one('div.point_num').text.strip()
else:
    score = ''
reg_date = target[0].select_one('dl.meta_item dd').text
if target[0].select_one('div.spec_list'):
    spec = target[0].select_one('div.spec_list').text.strip().replace('\n','').replace('\t', '')
else:
    spec = ''
print(title, URL, price, score, reg_date, spec)
```

```python
danawa = []
for i in target:
    try:
        title = i.select_one('img')['alt']
        URL = i.select_one('a')['href']
        price = i.select_one('p.price_sect').text.strip().replace('\n','').replace('가격정보 더보기','')
        if i.select_one('div.point_num'):
            score = i.select_one('div.point_num').text.strip()
        else:
            score = ''
        reg_date = i.select_one('dl.meta_item dd').text
        if i.select_one('div.spec_list'):
            spec = i.select_one('div.spec_list').text.strip().replace('\n','').replace('\t', '')
        else:
            spec = ''
        print(title, URL, price, score, reg_date, spec)
        danawa.append([title, URL, price, score, reg_date, spec])
    except:
        pass
```

```python
df = pd.DataFrame(danawa, columns=['제품명', '링크', '가격', '평점', '등록일', '스펙'])
df
```
