### CSV

```python
# 1) 관련 모듈 임포트 
import csv
import pandas as pd
import numpy as np
```

### with open() + csv.reader()
csv파일 => csv 객체 => 2차원 데이타 => 데이타프레임

```python
# 2) csv 파일 확인
# data/data.csv
ls data\data.csv
```

```python
# 3) csv파일 => csv 객체 => 리스트 저장
# 리스트 생성
data_list = []
with open('data/data.csv', 'r') as f:
    csv_data = csv.reader(f)
    print(csv_data)
    #  행단위로 출력후 리스트에 저장
    for row in csv_data:
        print(row)
        data_list.append(row)
```

```python
# 2차원 리스트
data_list
```

```python
# 4) 학생이름, 총점, 평균 형태로 출력
# 첫행은 타이틀이라서 제외
for c, name, kor, eng, mat, bio in data_list[1:]:
    tot = int(kor) + int(eng) + int(mat) + int(bio)
    avg = tot/4
#     print(name, tot, avg)
    print(f'{name:<10} 총점 : {tot} 평균 : {avg}')    
```

```python
# 5) 2차원리스트 => 데이타프레임화
df = pd.DataFrame(data_list[1:], columns=data_list[0])
df
```

### with open() + csv.reader()
- read_csv(csv 경로, encoding='utf-8/cp949'
- csv파일 =>  데이타프레임 => 넘파이배열 => 2차원리스트

```python
# 1) csv => 데이타프레임
df = pd.read_csv('data/data.csv')
df
```

```python
# 2) 데이타프레임 => 넘파이 배얼

arr = df.to_numpy()
arr
```

```python
# 3) 넘파이배열 => 2차원리스트
# 리스트명 = 넘파이배열명.tolist()
data_list = arr.tolist()
print(type(data_list))
data_list
```

```python
# 데이타프레임의 컬럼명만 출력 
df.columns
```

```python
# 4) 총점과 평균을 데이타로 추가해서 2차원리스트로 변형
# 빈리스트 생성
data_list2 = []
# 컬럼명만 첫행으로 추가
data_list2.append(['class', 'name', 'kor', 'eng', 'mat', 'bio', 'tot', 'avg'])
for c, name, kor, eng, mat, bio in data_list[1:]:
    tot = int(kor) + int(eng) + int(mat) + int(bio)
    avg = tot/4
    print(f'{name:<10} 총점 : {tot} 평균 : {avg}')  
    data_list2.append([c, name, kor, eng, mat, bio, tot, avg])
    
data_list2
```

```python
# 5) 데이타프레임화
df = pd.DataFrame(data_list2[1:], columns = data_list2[0])
df
```
