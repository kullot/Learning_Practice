```python
def book_api():
    word = input('책제목 입력하기 =>')
    # 네이버 검색 API예제는 블로그를 비롯 전문자료까지 호출방법이 동일하므로 blog검색만 대표로 예제를 올렸습니다.
    # 네이버 검색 Open API 예제 - 블로그 검색
    
    # 본인 아이디와 시크릿 키 설정 
    client_id = "Pk3ZiVWG_T3GtwePNpuN"
    client_secret = "cLQKT0MdNe"
    
    # 검색어 지정후 인코딩 
    encText = urllib.parse.quote(word)
    
    # url 생성 
    # 요청변수 파라미터 지정 
    # url 에 ?요청변수1=값1&요청변수2=값2
    # 반환되는 블로그 갯수 값 지정 
    # ?display=50
    # url = "https://openapi.naver.com/v1/search/blog?query=" + encText # json 결과
#     url = "https://openapi.naver.com/v1/search/blog?display=50&query=" + encText # json 결과
    # url = "https://openapi.naver.com/v1/search/blog.xml?query=" + encText # xml 결과
    url = "https://openapi.naver.com/v1/search/book.json?display=50&query=" + encText # json 결과
    
    print(url)
    
    # 페이지 요청 
    request = urllib.request.Request(url)
    request.add_header("X-Naver-Client-Id",client_id)
    request.add_header("X-Naver-Client-Secret",client_secret)
    response = urllib.request.urlopen(request)
    rescode = response.getcode()
    
    # 200 코드이면 response_body에 저장 
    if(rescode==200):
        response_body = response.read()
#         print(response_body.decode('utf-8'))
    else:
        print("Error Code:" + rescode)
    n_book = json.loads(response_body.decode('utf-8'))
    book_list = n_book['items']
    df = pd.DataFrame(book_list)
    n_list = []
    for item in book_list:
        title = item['title'].replace('<b>','').replace('</b>','')
        link =  item['link']
        image = item['image']
        author = item['author']
        price = item['price']
        discount = item['discount']
        publisher = item['publisher'].replace('<b>','').replace('</b>','')
        pubdate = item['pubdate']
        isbn = item['isbn']
        description = item['description'].replace('<b>','').replace('</b>','')
        n_list.append([title, link, image, author, price, discount, publisher, pubdate, isbn, description])
    df = pd.DataFrame(n_list, columns=['Title', 'Link', 'Image', 'Author', 'Price', 'Discount', 'Publisher', 'Pubdate', 'Isbn', 'Description'])
    return df
```
