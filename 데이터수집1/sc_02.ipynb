```python
# 0) 임포트
import requests
from bs4 import BeautifulSoup
```

```python
# 1) 웹페이지 접근
# 샘플 url => response 객체 => html 소스
# 예제 소스
# 국립고궁박물관 > 이용안내 > 관람안내 
url = 'http://www.gogung.go.kr/viewInfo.do'
res = requests.get(url)
# html 텍스트 소스
html_str = res.text
# html_str #가공되지 않은 Low Data
```

```python
# 2) BeautifulSoup 객체화
# soup변수명 = BeautifulSoup(html소스, 해석기(html.parser/html5/lxml/xml))
soup = BeautifulSoup(html_str, 'html.parser')
print(type(soup))
# soup
```

## BS4의 도트(.)을 이용한 태그 찾기

- 자식요소 찾기
- soup변수명.태그1.태그2...
- next_sibling : 형제 요소 정보 추출

- string/text : 태그안에 삽입된 텍스트 추출

- attrs : 속성 추출 => 딕셔너리, 태그명[속성]으로도 가능

```python
# 3) 도트를 이용한 태그요소 찾기

# h1 태그 찾기
print(soup.h1)
# h1 태그안의 a 태그 안의 img 찾기
print(soup.h1.a.img)
# h1 태그안의 a 태그 안의 img 태그의 태그이름, src 속성값 출력하기
print(soup.h1.a.img.name)
print(soup.h1.a.img['src'])
print(soup.h1.a.img.get('src'))
print('='*30)
# soup
```

BS4의 find 스타일의 메소드
- soup변수명.메소드(옵션)
- sfind(태그명) => 태그 1개
- sfind_all(태그명) => 태그 리스트
- sfind(id='아이디명') => 태그 1개
- sfind(class_='클래스명') => 태그 1개
- sfindall(class='클래스명') => 리스트
- find(attrs={속성:값}) => 태그 1개
- find_all(attrs={속성:값}) => 리스트

```python
# 4) find(), find_all() 태그요소 찾기
# 첫번째 img 태그 찾기
print(soup.find('img'))
# attrs 속성 => {속성:값 ...} 딕셔너리 형태
print(soup.find('img').attrs)
print(soup.find('img')['alt'])
print(soup.find('img')['src'])
# 이미지 주소를 웹상 절대경로로
base_url = 'http://www.gogung.go.kr'
print(base_url + soup.find('img')['src'])
```

```python
# 모든 img 테그를 찾아라
# find_all('테그명') => 리스트
img_list = soup.find_all('img')
print('몇개인가?',len(img_list))
print(type(img_list))
print('='*30)
# for 문을 이용하여 개별 요소 출력
for img in img_list:
    print(img)
print('='*30)
# for...in range() 문을 이용하여 개별 요소 출력
for i in range(len(img_list)):
    print(i, img_list[i])
# 모든 img 태그의 웹 절대 경로 출력
base_url = 'http://www.gogung.go.kr'
for img in img_list:
    print(base_url + img['src'])
```

```python
# 5) 특정 아이디로 요소 찾기
# find(id='아이디명') => 태그 1개
# <태그명 id='아이디명'> ....<태그>
# content 아이디 값으로 지정된 요소 찾기
# soup.find(id='content')

#아이디값이 content인 div 태그안의 모든 li를 찾아라
# div#content li
li_list = soup.find(id='content').find_all('li')
print(len(li_list))
#글자 부분만 출력
for li in li_list:
    print(li.text)
```

```python
# 6) 클래스명으로 찾기
# find(class_='클래스명') => 태그 1개
# find_all(class-='클래스명') => 리스트
# <태그명 class='클래스명1 클래스명2 ...'>....</태그>

# .section 클래스 찾기 - 1개만
print(soup.find(class_='section'))
# .section 클래스 안의 p 태그안의 strong 태그안에 삽입된 텍스트
print(soup.find(class_='section').p.strong.text)
print('='*50)
# .section 클래스 모두 찾기 => 리스트
section_list = soup.find_all(class_= 'section')ㄴ
print(len(section_list))

# 전시실 정규해설 내용 부분만 출력
print(section_list[3])
print('='*50)
# 전시실 정규해설 내용의 p부분만 출력
print(section_list[3].p.text)
```

```python
# 7) 특정 속성의 특정값으로 지정된 요소 찾기
# find(attr={속성:값}) => 태그 1개
# find_all(attrs={속성:값}) => 리스트
# find_all(태그명, attrs={속성:값}) => 리스트
# <태그 속성=값> ~ </태그>

# 모든 a 태그를 찾아라
print(soup.find_all('a'))
print('='*50)
# a 태그의 href 속성값이 # 인 요소를 찾아라 - 1개
print(soup.find(attrs={'href':'#'}))
print('='*50)
# a 태그의 href 속성값이 # 인 모든요소를 찾아라 - ?개
print(soup.find_all(attrs={'href':'#'}))
print('='*50)
a_list = soup.find_all(attrs={'href':'#'})
a_list = soup.find_all('a', attrs={'href':'#'})
print(len(a_list))
```

### BS4의 select 스타일의 메소드
- soup변수명.메소드(CSS선택자)
- select_one(CSS선택자) => 1개의 요소
- select(CSS선택자) => 리스트


### 태그
- select(태그명) : 모든 태그 요소를 찾아라 => 리스트   find_all(태그명)
- select_one(태그명) : 첫번째 태그 요소를 찾아라 => 1개만   find(태그명)

### 아이디
- select_one(#아이디명) : find(id=아이디명)

### 클래스
- select(. 클래스명) : find_all(class_=클래스명)
- select_one(. 클래스명) : find(class_=클래스명)

```python
# 8) 태그선택자로 찾기 - select_one(), select()
# 첫번째 img 태그를 찾아라
print(soup.select_one('img'))
print()
# 모든 img 태그를 찾아라
print(soup.select('img'))
```

```python
# css 선택자 스타일
# 태그1 태그2 ... : 태그1안의 삽입된 자손 요소
# 태그1 > 태그2 ... : 태그1안의 삽입된 직계 자식

# ul 태그안의 li 태그를 찾아라
print(len(soup.select('ul li')))
print(len(soup.select('ul > li')))

# ol 태그안의 li 태그를 찾아라
print('='*50)
print(len(soup.select('ol > li')))
print(len(soup.select('ol li')))
print('='*50)
```

```python
# tbody 태그안의 span 찾기
print(soup.select('tbody td'))
print(soup.select('tbody > tr > td > span'))
# AttributeError
# print(soup.select('tbody > tr > td > span').text)
print(soup.select('tbody > tr > td > span')[0].text)
print(soup.select_one('tbody > tr > td > span').text)
```

```python
# 9) 아이디 관련 선택자로 찾기 - select_one(), select()

# id값이 content
print(soup.select_one('content')) # None 태그로 찾기
print('='*50)
# print(soup.select_one('#content')) # 아이디로 찾기
# div#content => id값이 contentdls 인 div 태그
print(soup.select_one('div#content'))
print('='*50)
# 모든 p 태그를 찾아라
print(soup.select('p'))
print('='*50)
# 모든 content 아이디값을 가진 div 태그 안의 모든 div 태그를 찾아라
print(soup.select('div#content > div'))
```

```python
# 10) 클래스 관련 선택자로 찾기 - select_one(), select()


# assess 클래스값을 가진 div 태그를 찾아라
# select('태그명.클래스명 ...')
print(len(soup.select('div.assess')))
print(soup.select('div.assess'))
print('='*50)
# assess 클래스값을 가진 div 태그안의 star5 클래스값을 가진 span 태그를 찾아라
print(soup.select('div.assess span.star5'))
# assess 클래스값을 가진 div 태그안의 star5 클래스값을 가진 span 태그의 title 속성값을 찾아라
# TypeError 
print(soup.select('div.assess span.star5')['title'])
print('='*50)
print(soup.select('div.assess span.star5')[0]['title'])
```
