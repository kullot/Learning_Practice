4) 전체 남아수의 비율을 구하시요.

```python
txtRdd = sc.textFile('./data/births.txt')
```

```python
txtRdd.flatMap(lambda v:v.split(',')).collect()
```

```python
# 남아 수의 합과 평균
print('남아 수의 합:', txtRdd.map(lambda v: int(v.split(',')[1])).sum())
print('남아 수의 평균:', txtRdd.map(lambda v: int(v.split(',')[1])).mean())
```

```python
# 여아 수의 합과 평균
print('여아 수의 합:', txtRdd.map(lambda v: int(v.split(',')[2])).sum())
print('여아 수의 평균:', txtRdd.map(lambda v: int(v.split(',')[2])).mean())
```

```python
# 남아수가 높은 탑 5를 구하시오
print('top by year')
txtRdd.map(lambda v : v.split(', ')).map( lambda v : (v[0], v[1])).sortBy( lambda v : int(v[1]), ascending=False).take(5)
```

```python
# 전체 남아수의 비율을 구하시요.
txtRdd.map( lambda v: int(v.split(',')[1])).sum()/(txtRdd.map( lambda v: int(v.split(',')[1])).sum()+txtRdd.map( lambda v: int(v.split(',')[2])).sum())*100
```

```python
nBoy = txtRdd.map( lambda v: int(v.split(',')[1])).sum()
nGirl = txtRdd.map( lambda v: int(v.split(',')[2])).sum()
nBoy/(nBoy+nGirl)*100
```

```python
txtRdd.map( lambda v: int(v.split(',')[2])).histogram(10)
```

```python
# 여아수를 10개의 계급으로 했을때 각계급의 수를 구하시요
txtRdd.map( lambda v: int(v.split(',')[2])).histogram(10)
```

```python
TR = txtRdd.map( lambda v: int(v.split(',')[2])).histogram(10)
```

```python
# 여아수를 10개의 계급으로 했을때 각계급의 수를 구하시요
for i in range(len(TR[1])):
    print(TR[1][i])
```
